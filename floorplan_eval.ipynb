```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

def load_pgm(file_path):
    return cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

def display_image(img, title='Image', cmap='gray'):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()

def get_padding_and_crop(map1_shape, map2_shape, origin1, origin2, resolution):
    x1_pixel, y1_pixel = int(origin1[0] / resolution), int(origin1[1] / resolution)
    x2_pixel, y2_pixel = int(origin2[0] / resolution), int(origin2[1] / resolution)

    min_x = min(x1_pixel, x2_pixel)
    min_y = min(y1_pixel, y2_pixel)
    max_x = max(x1_pixel + map1_shape[1], x2_pixel + map2_shape[1])
    max_y = max(y1_pixel + map1_shape[0], y2_pixel + map2_shape[0])

    new_width = max_x - min_x
    new_height = max_y - min_y

    start_x1 = x1_pixel - min_x
    start_y1 = y1_pixel - min_y
    start_x2 = x2_pixel - min_x
    start_y2 = y2_pixel - min_y

    return new_height, new_width, start_x1, start_y1, start_x2, start_y2

def align_maps(map1, map2, origin1, origin2, resolution):
    new_height, new_width, start_x1, start_y1, start_x2, start_y2 = get_padding_and_crop(
        map1.shape, map2.shape, origin1, origin2, resolution
    )

    aligned_map1 = np.full((new_height, new_width), 255, dtype=np.uint8)
    aligned_map2 = np.full((new_height, new_width), 255, dtype=np.uint8)

    aligned_map1[start_y1:start_y1 + map1.shape[0], start_x1:start_x1 + map1.shape[1]] = map1
    aligned_map2[start_y2:start_y2 + map2.shape[0], start_x2:start_x2 + map2.shape[1]] = map2

    return aligned_map1, aligned_map2

def extract_obstacles(map_img):
    return np.where(map_img < 50, 0, 255)

def find_differences(map1, map2):
    return np.abs(map1.astype(np.int16) - map2.astype(np.int16))

def crop_to_region_of_interest(difference):
    non_zero_coords = np.argwhere(difference)
    top_left = non_zero_coords.min(axis=0)
    bottom_right = non_zero_coords.max(axis=0) + 1
    cropped_diff = difference[top_left[0]:bottom_right[0], top_left[1]:bottom_right[1]]
    return cropped_diff

def analyze_difference(map1_obstacles, map2_obstacles):
    black_pixels_map1 = (map1_obstacles == 0)
    black_pixels_map2 = (map2_obstacles == 0)
    
    # Only consider pixels that are black in both maps
    common_black_pixels = black_pixels_map1 | black_pixels_map2
    
    total_black_pixels = np.sum(common_black_pixels)
    different_black_pixels = np.sum(map1_obstacles[common_black_pixels] != map2_obstacles[common_black_pixels])
    
    similarity_percentage = ((total_black_pixels - different_black_pixels) / total_black_pixels) * 100
    print(f"Similarity: {similarity_percentage:.2f}%")

# Load the two PGM files
map1 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/test1/floorplan')
map2 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/gmap-test/gmap-test.pgm')

# Assume these origins are read from corresponding YAML files or given directly
origin1 = [-6.025000, -8.025000, 0.0]  # Example origin for map1
origin2 = [-100.000000, -102.60, 0.000000]  # Example origin for map2
resolution = 0.05  # Example resolution

# Align the maps
aligned_map1, aligned_map2 = align_maps(map1, map2, origin1, origin2, resolution)

# Display the aligned maps
display_image(aligned_map1, title='Aligned Map 1')
display_image(aligned_map2, title='Aligned Map 2')

# Extract obstacles from both maps
obstacles1 = extract_obstacles(aligned_map1)
obstacles2 = extract_obstacles(aligned_map2)

# Find differences in the black lines (obstacles) using numpy
differences = find_differences(obstacles1, obstacles2)

# Crop the differences to the region of interest
cropped_differences = crop_to_region_of_interest(differences)

# Display the cropped differences
display_image(cropped_differences, title='KdV vs Gmap', cmap='jet')

# Analyze the similarity
analyze_difference(obstacles1, obstacles2)
```
