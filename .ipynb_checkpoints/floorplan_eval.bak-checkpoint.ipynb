```python
import yaml
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def load_yaml(file_path):
    with open(file_path, 'r') as file:
        data = yaml.safe_load(file)
    return data

yaml1 = load_yaml("/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/test1/floorplan.yaml")
yaml2 = load_yaml("/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/test2/floorplan.yaml")

def load_pgm(file_path):
    image = Image.open(file_path)
    img_array = np.array(image)
    print(img_array.shape)
    return np.array(image)

pgm1 = load_pgm("/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/test1/floorplan")
pgm2 = load_pgm("/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/gmap-test/gmap-test.pgm")
print (pgm1, pgm2)

def compare_metadata(yaml1, yaml2):
    keys_to_compare = ['resolution', 'origin', 'occupied_thresh', 'free_thresh']
    differences = {}
    for key in keys_to_compare:
        if yaml1[key] != yaml2[key]:
            differences[key] = (yaml1[key], yaml2[key])
    return differences
    print (differences)

metadata_differences = compare_metadata(yaml1, yaml2)
if metadata_differences != 0:
    print("Metadata differences found:", metadata_differences)
else:
    print("Metadata is identical")

def pad_map(map1, map2):
    max_height = max(map1.shape[0], map2.shape[0])
    max_width = max(map1.shape[1], map2.shape[1])

    padded_map1 = np.zeros((max_height, max_width))
    padded_map2 = np.zeros((max_height, max_width))

    padded_map1[:map1.shape[0], :map1.shape[1]] = map1
    padded_map2[:map2.shape[0], :map2.shape[1]] = map2

    return padded_map1, padded_map2

pgm1, pgm2 = pad_map(pgm1, pgm2)
print (pgm1.shape, pgm2.shape)
print (pgm1, pgm2)

def overlay_maps(pgm1, pgm2):
    plt.figure(figsize=(10, 10))
    print("in")
    plt.imshow(pgm1, cmap='gray', alpha=0.5)
    plt.imshow(pgm2, cmap='hot', alpha=0.5)
    plt.title("Overlay of Map1 and Map2")
    plt.show()

overlay_maps(pgm1, pgm2)

def compare_maps(pgm1, pgm2):
    difference = pgm1 - pgm2
    plt.figure(figsize=(10, 10))
    plt.figure(difference, cmap='gray')
    plt.title("Difference between Map1 and Map2")
    plt.show()
    return difference

difference = compare_maps(pgm1, pgm2)

def analyze_difference(difference):
    total_pixels =  difference.size
    different_pixels = np.count_nonzero(difference)
    similarity_percentage = (1 - different_pixels / total_pixels) * 100
    print(f"Similarity: {similarity_percentage:.2f}%")

analyze_difference(difference)
    
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.metrics import mean_squared_error, structural_similarity

def load_pgm(file_path):
    return cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

def display_image(img, title='Image', cmap='gray'):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the two PGM files
map1 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/test1/floorplan')
map2 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/gmap-test/gmap-test.pgm')

# Display the original maps
display_image(map1, title='Map 1')
display_image(map2, title='Map 2')

# Initialize SIFT detector
sift = cv2.SIFT_create()

# Detect key points and descriptors
keypoints1, descriptors1 = sift.detectAndCompute(map1, None)
keypoints2, descriptors2 = sift.detectAndCompute(map2, None)

# Initialize BFMatcher
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

# Match descriptors
matches = bf.match(descriptors1, descriptors2)

# Sort matches by distance
matches = sorted(matches, key=lambda x: x.distance)

# Draw matches (optional)
img_matches = cv2.drawMatches(map1, keypoints1, map2, keypoints2, matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
display_image(img_matches, title='Feature Matches')

# Extract location of good matches
src_pts = np.float32([keypoints1[m.queryIdx].pt for m in matches]).reshape(-1, 1, 2)
dst_pts = np.float32([keypoints2[m.trainIdx].pt for m in matches]).reshape(-1, 1, 2)

# Estimate transformation matrix using RANSAC
M, mask = cv2.estimateAffinePartial2D(src_pts, dst_pts, method=cv2.RANSAC)

print("Transformation Matrix:\n", M)

# Apply the transformation
rows, cols = map1.shape
aligned_map2 = cv2.warpAffine(map2, M, (cols, rows))

# Display the aligned map
display_image(aligned_map2, title='Aligned Map 2')

# Resize the aligned map to match the reference map size
aligned_map2_resized = cv2.resize(aligned_map2, (map1.shape[1], map1.shape[0]))

# Perform numerical comparisons
mse = mean_squared_error(map1, aligned_map2_resized)
ssim, _ = structural_similarity(map1, aligned_map2_resized, full=True)

print(f'Mean Squared Error: {mse}')
print(f'Structural Similarity Index: {ssim}')

# Display the comparison result
difference = map1 - aligned_map2_resized
display_image(difference, title='Difference Image', cmap='jet')

def analyze_difference(difference):
    total_pixels =  difference.size
    different_pixels = np.count_nonzero(difference)
    similarity_percentage = (1 - different_pixels / total_pixels) * 100
    print(f"Similarity: {similarity_percentage:.2f}%")

analyze_difference(difference)
```

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

def load_pgm(file_path):
    return cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

def display_image(img, title='Image', cmap='gray'):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()

def get_padding_and_crop(map1_shape, map2_shape, origin1, origin2, resolution):
    x1_pixel, y1_pixel = int(origin1[0] / resolution), int(origin1[1] / resolution)
    x2_pixel, y2_pixel = int(origin2[0] / resolution), int(origin2[1] / resolution)

    min_x = min(x1_pixel, x2_pixel)
    min_y = min(y1_pixel, y2_pixel)
    max_x = max(x1_pixel + map1_shape[1], x2_pixel + map2_shape[1])
    max_y = max(y1_pixel + map1_shape[0], y2_pixel + map2_shape[0])

    new_width = max_x - min_x
    new_height = max_y - min_y

    start_x1 = x1_pixel - min_x
    start_y1 = y1_pixel - min_y
    start_x2 = x2_pixel - min_x
    start_y2 = y2_pixel - min_y

    return new_height, new_width, start_x1, start_y1, start_x2, start_y2

def align_maps(map1, map2, origin1, origin2, resolution):
    new_height, new_width, start_x1, start_y1, start_x2, start_y2 = get_padding_and_crop(
        map1.shape, map2.shape, origin1, origin2, resolution
    )

    aligned_map1 = np.full((new_height, new_width), 255, dtype=np.uint8)
    aligned_map2 = np.full((new_height, new_width), 255, dtype=np.uint8)

    aligned_map1[start_y1:start_y1 + map1.shape[0], start_x1:start_x1 + map1.shape[1]] = map1
    aligned_map2[start_y2:start_y2 + map2.shape[0], start_x2:start_x2 + map2.shape[1]] = map2

    return aligned_map1, aligned_map2

def extract_obstacles(map_img):
    return np.where(map_img < 50, 0, 255)

def find_differences(map1, map2):
    return np.abs(map1.astype(np.int16) - map2.astype(np.int16))

def crop_to_region_of_interest(difference):
    non_zero_coords = np.argwhere(difference)
    top_left = non_zero_coords.min(axis=0)
    bottom_right = non_zero_coords.max(axis=0) + 1
    cropped_diff = difference[top_left[0]:bottom_right[0], top_left[1]:bottom_right[1]]
    return cropped_diff

# Load the two PGM files
map1 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/test1/floorplan')
map2 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/gmap-test/gmap-test.pgm')

# Assume these origins are read from corresponding YAML files or given directly
origin1 = [-6.025000, -8.025000, 0.0]  # Example origin for map1
origin2 = [-100.000000, -100.000000, 0.000000]  # Example origin for map2
resolution = 0.05  # Example resolution

# Align the maps
aligned_map1, aligned_map2 = align_maps(map1, map2, origin1, origin2, resolution)

# Display the aligned maps
display_image(aligned_map1, title='Aligned Map 1')
display_image(aligned_map2, title='Aligned Map 2')

# Extract obstacles from both maps
obstacles1 = extract_obstacles(aligned_map1)
obstacles2 = extract_obstacles(aligned_map2)

# Find differences in the black lines (obstacles) using numpy
differences = find_differences(obstacles1, obstacles2)

# Crop the differences to the region of interest
cropped_differences = crop_to_region_of_interest(differences)

# Display the cropped differences
display_image(cropped_differences, title='Differences in Obstacles (Zoomed)', cmap='jet')

def analyze_difference(difference):
    total_pixels = difference.size
    different_pixels = np.count_nonzero(difference)
    similarity_percentage = (1 - different_pixels / total_pixels) * 100
    print(f"Similarity: {similarity_percentage:.2f}%")

analyze_difference(cropped_differences)
```

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

def load_pgm(file_path):
    return cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

def display_image(img, title='Image', cmap='gray'):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()

def get_padding_and_crop(map1_shape, map2_shape, origin1, origin2, resolution):
    x1_pixel, y1_pixel = int(origin1[0] / resolution), int(origin1[1] / resolution)
    x2_pixel, y2_pixel = int(origin2[0] / resolution), int(origin2[1] / resolution)

    min_x = min(x1_pixel, x2_pixel)
    min_y = min(y1_pixel, y2_pixel)
    max_x = max(x1_pixel + map1_shape[1], x2_pixel + map2_shape[1])
    max_y = max(y1_pixel + map1_shape[0], y2_pixel + map2_shape[0])

    new_width = max_x - min_x
    new_height = max_y - min_y

    start_x1 = x1_pixel - min_x
    start_y1 = y1_pixel - min_y
    start_x2 = x2_pixel - min_x
    start_y2 = y2_pixel - min_y

    return new_height, new_width, start_x1, start_y1, start_x2, start_y2

def align_maps(map1, map2, origin1, origin2, resolution):
    new_height, new_width, start_x1, start_y1, start_x2, start_y2 = get_padding_and_crop(
        map1.shape, map2.shape, origin1, origin2, resolution
    )

    aligned_map1 = np.full((new_height, new_width), 255, dtype=np.uint8)
    aligned_map2 = np.full((new_height, new_width), 255, dtype=np.uint8)

    aligned_map1[start_y1:start_y1 + map1.shape[0], start_x1:start_x1 + map1.shape[1]] = map1
    aligned_map2[start_y2:start_y2 + map2.shape[0], start_x2:start_x2 + map2.shape[1]] = map2

    return aligned_map1, aligned_map2

def extract_obstacles(map_img):
    return np.where(map_img < 50, 0, 255)

def find_differences(map1, map2):
    return np.abs(map1.astype(np.int16) - map2.astype(np.int16))

def crop_to_region_of_interest(difference):
    non_zero_coords = np.argwhere(difference)
    top_left = non_zero_coords.min(axis=0)
    bottom_right = non_zero_coords.max(axis=0) + 1
    cropped_diff = difference[top_left[0]:bottom_right[0], top_left[1]:bottom_right[1]]
    return cropped_diff

def analyze_difference(map1_obstacles, map2_obstacles):
    black_pixels_map1 = (map1_obstacles == 0)
    black_pixels_map2 = (map2_obstacles == 0)
    
    # Only consider pixels that are black in both maps
    common_black_pixels = black_pixels_map1 | black_pixels_map2
    
    total_black_pixels = np.sum(common_black_pixels)
    different_black_pixels = np.sum(map1_obstacles[common_black_pixels] != map2_obstacles[common_black_pixels])
    
    similarity_percentage = ((total_black_pixels - different_black_pixels) / total_black_pixels) * 100
    print(f"Similarity: {similarity_percentage:.2f}%")

# Load the two PGM files
map1 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/test1/floorplan')
map2 = load_pgm('/home/user/Downloads/KdRobotDockers-main/Gemini_ros_noetic/robot_ws/2024_07_04/gmap-test/gmap-test.pgm')

# Assume these origins are read from corresponding YAML files or given directly
origin1 = [-6.025000, -8.025000, 0.0]  # Example origin for map1
origin2 = [-100.000000, -102.60, 0.000000]  # Example origin for map2
resolution = 0.05  # Example resolution

# Align the maps
aligned_map1, aligned_map2 = align_maps(map1, map2, origin1, origin2, resolution)

# Display the aligned maps
display_image(aligned_map1, title='Aligned Map 1')
display_image(aligned_map2, title='Aligned Map 2')

# Extract obstacles from both maps
obstacles1 = extract_obstacles(aligned_map1)
obstacles2 = extract_obstacles(aligned_map2)

# Find differences in the black lines (obstacles) using numpy
differences = find_differences(obstacles1, obstacles2)

# Crop the differences to the region of interest
cropped_differences = crop_to_region_of_interest(differences)

# Display the cropped differences
display_image(cropped_differences, title='KdV vs Gmap', cmap='jet')

# Analyze the similarity
analyze_difference(obstacles1, obstacles2)
```
